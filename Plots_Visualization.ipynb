Scatter Plot using Matplot library

```python
%matplotlib inline

from azureml import Workspace
import pandas as pd
ws = Workspace()
ds = ws.datasets['breast-cancer-wisconsin.csv']
data_frame = ds.to_dataframe()
data_frame.plot.scatter(x=1,y=0)
```

Pairwise Plot using seaborn library

```python
%matplotlib inline

from azureml import Workspace
import pandas as pd
import seaborn as sns

ws = Workspace()
ds = ws.datasets['breast-cancer-wisconsin.csv']
data_frame = ds.to_dataframe()

# Pai-wise plot
sns.pairplot(data_frame)
```

```python
%matplotlib inline

from azureml import Workspace
import pandas as pd
import seaborn as sns

ws = Workspace()
ds = ws.datasets['breast-cancer-wisconsin.csv']
data_frame = ds.to_dataframe()
data_frame.columns = ['ID','radius','texture','perimeter','area','smoothness','compactness','concavity','concave','symmetry','fractal_dim']
x=data_frame['radius']
x.plot(kind='hist')
```

```python
%matplotlib inline

from azureml import Workspace
import pandas as pd
import matplotlib.pyplot as plt

ws = Workspace()
ds = ws.datasets['breast-cancer-wisconsin.csv']
data_frame = ds.to_dataframe()
data_frame.columns = ['ID','radius','texture','perimeter','area','smoothness','compactness','concavity','concave','symmetry','fractal_dim']
plt.boxplot(data_frame['radius'])
```
